2) Write a script or function that does the following: (i) it takes two numbers (x_0 and r) as input; (ii) then, it iteratively calls function y = f(r,x) in a loop, such that the value of y returned in each iteration is used as the new value of x for the next iteration. The loop should include 50 iterations, starting with x = x_0. [Note: set the number of iterations through a predefined or externally provided variable (i.e., first define n_iter = 50; then run n_iter iterations); that will make it easier to reuse the code in future exercises]. (iii) At each iteration, record the value of x. (iv) After the last iteration, plot the series of values of x (the values of x should go in the vertical axis and the iteration number in the horizontal axis). If you find plotting the data challenging, you can just save the values of x in a text file and plot them with excel or any plotting software (e.g. xmgrace). The plots should look more or less like these ones.

```python
from matplotlib import pyplot as plt

x_0 = 0.7
r = 3
n_iter = 50

def plot_x_values(x_0,r,n_iter, save_figs=False):
    
    x_values = []
    x = x_0

    for i in range(n_iter):
        x_values.append(x)
        y = r*x*(1-x)
        x = float(y)

    fig, ax = plt.subplots()
    ax.plot(list(range(n_iter)), x_values)
    ax.set_xlim((0,50))
    ax.set_xlabel("Iterations")
    ax.set_ylabel("x value")
    ax.set_title("r: {}   x: {}".format(r,x_0))

plot_x_values(x_0,r,n_iter)
```

4) Run the script with x_0 = 0.5 and different values of r. Try at least: r = 1, r = 2.5, r = 3.2, r = 3.5, r = 3.8, r = 5. What happens if you start from a slightly different x_0 (e.g. 0.4)?

```python
r_list = [1, 2.5, 3.2, 3.5, 3.8, 5]
x_0 = 0.5

for r in r_list:
    plot_x_values(x_0, r, n_iter)
    plt.savefig(fname = "my_plot_{}.jpg".format(r_list.index(r)+1), dpi = 200)
```
